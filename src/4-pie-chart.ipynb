### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

# Data for plotting
labels = ['Python', 'Java', 'C++', 'JavaScript']  # Programming languages
sizes = [40, 25, 20, 15]  # Corresponding market share sizes

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Customize the plot
plt.title("Programming Language Market Share")

# Display the plot
plt.show()  # Show the pie chart
```
